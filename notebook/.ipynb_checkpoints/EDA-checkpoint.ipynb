## Create functions for model training and evaluation

```python
def evaluate_models(X,y,models):
    '''
    This function takes in X and y and models dictionary as input
    It splits the data into Train Test split
    Iterates through the given model dictionary and evaluates the metrics
    Returns: Dataframe which contains report of all models metrics
    '''
    exclude_scaling=["Decsion Tree","Random Forest","Gradient Boost","Ada Boost","XgBoost"]
    xtrain,xtest,ytrain,ytest = train_test_split(X,y,random_state=42,test_size=0.2)
    scaler = StandardScaler()
    r2_score_list =[]
    adjusted_r2_score = []
    mae =[]
    mse = []
    rmse = []
    for model in models:
        if model not in exclude_scaling:
            xtrain = scaler.fit_transform(xtrain)
            xtest = scaler.transform(xtest)
        temp_model = models[model]
        print(temp_model)
        temp_model.fit(xtrain,ytrain)
        y_pred = temp_model.predict(xtest)
        temp_r2_score = r2_score(ytest,y_pred)
        r2_score_list.append(temp_r2_score)
        mae.append(mean_absolute_error(ytest,y_pred))
        mse.append(mean_squared_error(ytest,y_pred))
        rmse.append(np.sqrt(mean_squared_error(ytest,y_pred)))
        adjustedr2 = 1-(((1-temp_r2_score)*(X.shape[0]-1))/(X.shape[0]-len(X.columns)+1-1))
        adjusted_r2_score.append(adjustedr2)
       
    report_df = pd.DataFrame([r2_score_list,adjusted_r2_score,mae,mse,rmse],index=["R2 Score","Adjusted R2 Score","MAE","MSE","RMSE"])
    report_df.columns = models.keys()
    return report_df.T.sort_values(by=["R2 Score","Adjusted R2 Score"],ascending=False)
        
    
    
```

## Experiment 1. Model Training without handling outliers

#### Splitting dependent indendent features

```python
X=df.drop([df.columns[-1]],axis=1)
y=df[df.columns[-1]]
```

```python
models = {
    "Linear Regression":LinearRegression(),
    "Ridge Regression":Ridge(),
    "Lasso Regression":Lasso(),
    "ElasticNet Rgression":ElasticNet(),
    "Decsion Tree":DecisionTreeRegressor(),
    "Random Forest":RandomForestRegressor(),
    "Gradient Boost":GradientBoostingRegressor(),
    "Ada Boost":AdaBoostRegressor(),
    "XgBoost":XGBRegressor(learning_rate=0.15,n_estimators=150),
}
```

```python
report =evaluate_models(X,y,models)
```

### Experiment 2. After handling Outliers

```python
plt.figure(figsize=(15,10))
plt.suptitle("Univariate analysis of Features")
for i in range(len(df.columns)):
    plt.subplot(3,3,i+1)
    sns.boxplot(data=df[df.columns[i]])
    plt.xlabel(df.columns[i])
    plt.ylabel("Density")
    plt.tight_layout()
```

#### <b>Observations: </b>
* Outliers present in below columns:
    * Blast Furnace
    * Water
    * Super Plasticizer
    * Age
    * Concrete
* We have to handle the outliers for better performance of model

```python
def find_boundaries(df,variable,distance):
    IQR = df[variable].quantile(0.75)-df[variable].quantile(0.25)
    lower_boundary = df[variable].quantile(0.25) - (IQR * distance)
    upper_boundary = df[variable].quantile(0.75) + (IQR * distance)
    return upper_boundary,lower_boundary
```

```python
def features_with_outliers(df):
    outliers_features = {}
    for i in range(len(df.columns)-1):
        upper_limit,lower_limit = find_boundaries(df,df.columns[i],1)
        outliers_features[df.columns[i]] = np.where(df[df.columns[i]]>upper_limit,True,
                                                np.where(df[df.columns[i]] < lower_limit, True, False))
    return outliers_features
```

```python
def remove_outliers(df):
    out = features_with_outliers(df)
    for i in out.keys():
        df_trimmed = df.loc[~out[i] ]
    return df_trimmed
```

```python
df_trimmed = remove_outliers(df)
```
